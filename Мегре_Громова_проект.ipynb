```python
# получаем API ключи
try:
    GROQ_API_KEY = userdata.get('GROQ_API_KEY')
except:
    GROQ_API_KEY = os.getenv("GROQ_API_KEY")

try:
    TAVILY_API_KEY = userdata.get('TAVILY_API_KEY')
except:
    TAVILY_API_KEY = os.getenv("TAVILY_API_KEY")
```

### 1.1. Инструменты built-in библиотеки smolagents

```python
# Попробуем применить DuckDuckGoSearchTool для поиска в интернете

# тестовый запрос
input_text = "Найди 3 идеи для рилсов для блога о саморазвитии"

# создаем модель LiteLLM для работы с Groq
model = LiteLLMModel(
    model_id="groq/llama-3.1-8b-instant",
    api_key=GROQ_API_KEY,
    temperature=0.2
)

# создаем агента с веб-поиском
agent = CodeAgent(
    tools=[DuckDuckGoSearchTool()],
    model=model,
    max_steps=2,
    verbosity_level=1
)

# запускаем агента
print(input_text)
print("-" * 50)

result = agent.run(input_text)
print(result)
```

```python
# Инструмент VisitWebpageTool для чтения веб-страниц

visit_tool = VisitWebpageTool()

print("Тест VisitWebpageTool")
page_text = visit_tool.forward("https://ru.wikipedia.org/wiki/Контент-маркетинг")
print(page_text[:1000])
```

```python
# Протестируем также инструмент PythonInterpreterTool, который проверяет можно ли выполнять Python-код внутри агента

python_tool = PythonInterpreterTool()

print("Тест PythonInterpreterTool")
python_result = python_tool.forward("""
titles = ["Идея для Reels", "Контент-план на неделю", "Как оформить профиль блогера"]
lengths = [len(x) for x in titles]
result = lengths
result
""")
print(python_result)
```

### 1.2. Кастомные инструменты библиотеки smolagents

```python
# Пример инструмента для посчета длины текста

class WordCountTool(Tool):
    name = "word_count"
    description = """
    Подсчитывает количество слов и символов в тексте.
    Используй когда нужно узнать длину текста или статьи.
    """
    inputs = {
        "text": {"type": "string", "description": "Текст для подсчёта"}
    }
    output_type = "string"

    def forward(self, text):
        words = len(text.split())
        chars = len(text)
        return f"Слов: {words}\nСимволов: {chars}"

word_count_tool = WordCountTool()

print("\nТест WordCountTool")
print(word_count_tool.forward("Пост для блогера должен быть полезным, понятным и цепляющим внимание аудитории."))


@tool
def content_plan_generator(topic: str, days: int, ideas: str) -> str:
    """
    Создает простой контент-план для блогера.

    Args:
        topic: Тема блога.
        days: Количество дней в контент-плане.
        ideas: Идеи для контента, переданные одной строкой через символ ;.

    Returns:
        Готовый контент-план в виде строки.
    """
    ideas_list = [idea.strip() for idea in ideas.split(";")]
    plan = [f"Контент-план по теме: {topic}"]
    for i in range(min(days, len(ideas_list))):
        plan.append(f"День {i+1}: {ideas_list[i]}")
    return "\n".join(plan)

print("\nТест content_plan_generator")
print(content_plan_generator(
    topic="личный бренд",
    days=3,
    ideas="Рилс: 3 ошибки начинающего блогера; Пост: как оформить шапку профиля; Рилс: идеи сторис на каждый день"
))

```

### 1.3. Инструменты built-in библиотеки langgraph

### 1.4. Кастомные инструменты библиотеки langgraph

```python
# кастомный инструмент для подсчета слов

@lc_tool
def lc_word_count(text: str) -> str:
    """Counts words in a given text. Use when you need to know the length of an article."""
    return f"Words: {len(text.split())}"

print("Тест кастомного langgraph-инструмента lc_word_count")
print(lc_word_count.invoke({"text": "нанотехнологии это круто"}))
```

```python
# Кастомный инструмент для создания контент-плана по теме

@lc_tool
def make_content_plan(topic: str) -> str:
    """Creates a simple 5-day content plan for a blogger based on the topic."""
    return (
        f"5-дневный контент-план по теме '{topic}':\n"
        "День 1 — Пост: полезный разбор темы\n"
        "День 2 — Рилс: 3 быстрых совета\n"
        "День 3 — Пост: личный опыт / кейс\n"
        "День 4 — Рилс: частые ошибки новичков\n"
        "День 5 — Пост: призыв к обсуждению с подписчиками"
    )

print("Тест кастомного langgraph-инструмента make_content_plan")
print(make_content_plan.invoke({"topic": "личный бренд"}))
```

## 2. Реализация логики агента

Для реализации агентной логики была выбрана библиотека smolagents, так как она предоставляет более простой и быстрый способ создания агента с поддержкой инструментов без необходимости проектирования сложной структуры графа. Кроме того, в отличие от langgraph, который ориентирован на более сложные сценарии, smolagents лучше подходит для решения задач с линейной последовательностью шагов, что соответствует целям данного проекта.

Для функционала агента мы выбрали 3 инструмента:

1. DuckDuckGoSearchTool для поиска в интернете информации
2. Кастомный инструмент content_plan_generator для генерации контент-плана
3. Кастомный инструмент WordCountTool для подсчета символов в тексте


**Настройка параметров max_steps, verbosity_level**

Наш агент предназначит для помощи блогеру в генерации контента и решает 3 задачи: генерирует идеи публикаций через поиск в интернете, считает символы в текстах для публикации и генерирует контент-план на заданый промежуток времени. Параметр max_steps определяет сколько шагов делает агент. Чем больше значение этого параметра, тем более гибким становится ответ агента, однако возможно больше лишней информации. Поскольку функционал нашего агента ограничен 3 инструментами, мы остановились на значении max_steps=5. Параметр verbosity_level влияет на уровень подробности промежуточного вывода агента. В данном случае мы выбрали значение verbosity_level=1, чтобы агент также выводил служебную информацию. При таком значении параметра, модель выводит резултаты поиска в интернете инструмента DuckDuckGoSearchTool и пользователь может посмотреть источники, проверить их качество и перейти по ссылки для более детального изучения. Кроме того, это позволит нам более точно проанализировать возможные ошибки агента.

```python
# Собираем агента

model = LiteLLMModel(
    model_id="groq/llama-3.1-8b-instant",
    api_key=GROQ_API_KEY,
    temperature=0.7
)

# все инструменты
tools = [
    DuckDuckGoSearchTool(), # built-in
    word_count_tool, # кастомный
    content_plan_generator # кастомный
]

agent = CodeAgent(
    tools=tools,
    model=model,
    max_steps=5,
    verbosity_level=1
)

```

## 3. Тестирование и анализ

```python
# Тест 1
query = (
    """
    Найди идеи для контента по теме личного бренда.
    На основе найденной информации составь 5-дневный контент-план.
    Затем посчитай количество слов и символов в этом контент-плане.
    Верни финальный результат.
    """
)

print(f"Запрос: {query}")
print("-" * 60)

result = agent.run(query)

print("-" * 60)
print("Итоговый ответ агента:")
print(result)
```

Тест можно считать частично успешным. Агент корректно разбил задачу на шаги и последовательно вызвал инструменты. Сначала был выполнен поиск, затем вызван инструмент генерации контент-плана, после чего агент попытался применить инструмент подсчета слов и символов. Однако на этапе формирования финального ответа возникла ошибка, так как WordCountTool возвращает строку, а агент попытался обратиться к результату как к словарю. Кроме того, контент-план получился некачественным, поскольку в него была включена сырая поисковая выдача

```python
# Тест 2
query = (
    """
    Найди в интернете идеи для reels о саморазвитии,
    выбери наиболее подходящую тему,
    составь по ней контент-план и посчитай число слов в плане."
    """
)

print(f"Запрос: {query}")
print("-" * 60)

result = agent.run(query)

print("-" * 60)
print("Итоговый ответ агента:")
print(result)
```

В тесте 2 агент продемонстрировал частично успешное выполнение многошагового запроса: он корректно начал с интернет-поиска идей для reels о саморазвитии, затем попытался выбрать тему, сгенерировать контент-план и подсчитать количество слов, то есть разбил задачу на последовательные этапы и использовал несколько инструментов. Однако качество выполнения оказалось снижено из-за ошибок на этапах интерпретации и использования инструментов: агент выбрал тему «Финансовая грамотность», которая лишь косвенно связана с запросом о саморазвитии. После исправления этой ошибки агент смог подсчитать слова, но финальный ответ оказался слишком кратким и неполным, так как не включал сам контент-план и не пояснял, почему была выбрана именно эта тема. Таким образом, тест показывает, что агент способен выполнять многошаговые действия последовательно, но итоговое решение следует оценить как частично успешное из-за слабой релевантности выбора темы, промежуточной ошибки в коде и недостаточно полного финального ответа.

```python
# Тест 3
query = (
    """
    Найди в интернете идеи для постов в блоге про путешествия.
    Выбери 3 наиболее интересных тем,
    Составь из этих тем контент-план на 3 дня.
    """
)

print(f"Запрос: {query}")
print("-" * 60)

result = agent.run(query)

print("-" * 60)
print("Итоговый ответ агента:")
print(result)
```

В результате после нескольких повторных шагов контекст выполнения стал слишком большим, и модель Groq вернула ошибку превышения лимита токенов. Таким образом, агент не довёл выполнение задачи до конца, а причина неуспешности связана с ошибочной промежуточной логикой и накоплением слишком большого объема служебного контекста.
