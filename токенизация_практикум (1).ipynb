
# Токенизация в NLP

## Введение

**Токенизация** – это процесс разбиения текста на минимальные единицы (токены): слова, подслова или символы.  
Она является первым шагом во многих NLP задачах: анализ тональности, машинный перевод, чат-боты, обучение языковых моделей.

В этом туториале мы разберём разные виды токенизации и библиотеки, которые их реализуют.



## 1. Простые методы токенизации

### 1.1. Разделение по пробелам


```python
text = "Hello, world! This is a test."
text.split()
```


### 1.2. Регулярные выражения


```python
import re

text = "Hello, world! This is a test."
tokens = re.findall(r"\w+", text)
tokens
```


**Ограничения простых методов**:  
- Не учитывают пунктуацию.  
- Не работают с разными языками и морфологией.  
- Не подходят для серьёзных задач.  



## 2. Токенизация с помощью библиотек



### 2.1. NLTK  
**Применение**: образовательные проекты, простая предобработка текста.  
**Ограничения**: работает медленно на больших корпусах.  
**Когда использовать**: для курсовых и лабораторных работ по основам NLP.


```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')

from nltk.tokenize import word_tokenize, sent_tokenize

text = "Hello, world! This is a test."
print(word_tokenize(text))
print(sent_tokenize(text))
```


### 2.2. spaCy  
**Применение**: промышленные проекты, морфология, синтаксис.  
**Ограничения**: большие модели, требует ресурсов.  
**Когда использовать**: для курсовых по синтаксису и морфологии, проектов по обработке естественного языка.


```python
import spacy

nlp = spacy.load("en_core_web_sm")
doc = nlp("Hello, world! This is a test.")
[t.text for t in doc]
```


### 2.3. Stanza  
**Применение**: многоязычные корпуса, морфология, синтаксис.  
**Ограничения**: медленнее spaCy, требует загрузки моделей.  
**Когда использовать**: в исследованиях и при работе с редкими языками.


```python
!pip install stanza
import stanza

stanza.download("en")
nlp = stanza.Pipeline("en")
doc = nlp("Hello, world! This is a test.")
[[word.text for word in sent.words] for sent in doc.sentences]
```


### 2.4. MosesTokenizer (sacremoses)  
**Применение**: машинный перевод, предобработка текстов.  
**Ограничения**: устаревающий стандарт.  
**Когда использовать**: для экспериментов с классическими MT-системами.


```python
!pip install sacremoses

from sacremoses import MosesTokenizer

mt = MosesTokenizer()
mt.tokenize("Hello, world! This is a test.")
```


### 2.5. gensim  
**Применение**: подготовка данных для моделей `word2vec`, `fastText`.  
**Ограничения**: базовая токенизация.  
**Когда использовать**: при обучении собственных векторных моделей.  


```python
!pip install gensim
from gensim.utils import simple_preprocess

simple_preprocess("Hello, world! This is a test.")
```


### 2.6. HuggingFace Tokenizers  
**Применение**: трансформеры, нейросетевые модели.  
**Ограничения**: сложность, нужен GPU для обучения токенизаторов.  
**Когда использовать**: в курсовых и проектах с BERT, GPT, T5.


```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
tokens = tokenizer("Hello, world! This is a test.")
tokens
```

## 5. Сравнение токенизаторов  

Попробуем применить разные методы к одному и тому же тексту.  



## 6. Практика

Возьмём новостной текст, токенизируем разными способами и сравним.  


```python
# выведем все импорты в отдельную ячейку

# NLTK
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize, sent_tokenize

# SpaCy
import spacy
nlp = spacy.load("en_core_web_sm")

# Gensim
!pip install gensim
from gensim.utils import simple_preprocess
```

```python
article = 'Natural language processing (NLP) is a field of artificial intelligence\
that gives computers the ability to understand text and spoken words in much the same way human beings can.'

"""
NLTK
"""
print('Токенизация текста с помощью NLTK:')
print(word_tokenize(article))

"""
SpaCy
"""
doc = nlp(article)
print('\nТокенизация текста с помощью SpaCy:')
print([t.text for t in doc])

"""
Gensim
"""
print('\nТокенизация текста с помощью Gensim:')
simple_preprocess(article)
```

В этом задании сравнивались методы токенизации текста, реализованные с помощью 3 разных библиотек: NLTK, SpaCy, Gensim. Сравнение результатов показывает, что токенизация с помощью NLTK и SpaCy, достаточно точно сохраняет структуру исходного текста: сохранен регистр, выделены знаки препинания в отдельные токены. При этом оба эти метода чувствительны к качеству исходных данных. Так, если в исходном тексте слово разделено некорректно,оно сохраняет исходную запись  (пример из задания – intelligencethat). В отличие от NLTK и SpaCy, токенизация с помощью Gensim более нормализированная: все токены приводятся к нижнему регистру, пунктуация удаляется, сохраняются только содержательные слова. Так, токенизация с помощью Gensim хорошо может подходить для задач тематического моделирования и векторизации, но может также приводить к потере части лингвистической информации и потенциально значимых элементов исходного текста. Таким образом, рассмотренные варианты токенизации отличаются и подходят для разных задач, в заивисимости от цели исследования.

Термин Out-of-Vocabulary обозначает проблему ограниченных словарей, при которой слова или токены, отсутствующие в словаре модели, сформированном на этапе обучения, заменяются на <unk> и не могут быть корректно обработаны моделью. В контексте машинного перевода это приводит к тому, что редкие, новые или морфологически сложные формы не могут быть корректно обработаны,что приводит к потере семантической и грамматической информации.

2. В статье “Achieving Open Vocabulary Neural Machine Translation with Hybrid Word-Character Models” авторы предлагают гибридную архитектуру, сочетающую обработку на уровне слов и символов, что позволяет создать так называемый "открытый словарь". Основная идея заключается в том, что для частотных слов используется стандартное word-level представление, тогда как редкие или неизвестные слова кодируются с помощью символьной модели: их векторные представления строятся на основе последовательности символов, а на стороне декодера такие слова могут генерироваться посимвольно вместо замены на <unk>. Благодаря этому подходу модель сохраняет эффективность word-level NMT и получает способность корректно обрабатывать ранее невидимые словоформы, что значительно решает проблему Out-of-Vocabulary и повышает качество перевода.
